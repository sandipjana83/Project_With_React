```python
from sympy import symbols,integrate
x,y=symbols('x,y')
exp=3*x**2+4*x+5
integration=integrate(exp,x)
print(integration)
```

```python
from sympy import symbols,integrate,sin,cos
x,y=symbols('x,y')
exp=sin(x)*cos(x)
integration=integrate(exp,x)
print(integration)
```

```python
from sympy import symbols,integrate
x,y=symbols('x,y')
exp=x**2+1
integration=integrate(exp,(x,0,2))
print(integration)
```

```python
from sympy import symbols,integrate,sin,pi
x,y=symbols('x,y')
exp=sin(x)
integration=integrate(exp,(x,0,pi))
print(integration)
```

```python
from sympy import symbols,solve
x=symbols('x')
equation = 3*x + 7 - (2*x - 5)
solution=solve(equation,x)
print(solution)
```

```python
from sympy import symbols,solve
x=symbols('x')
equation =(4*x+5)-(x-3)
solution=solve(equation,x)
print(solution)
```

```python
from sympy import symbols,solve,Eq
x,y=symbols('x,y')
eq1=Eq(2*x+y,7)
eq2=Eq(x-y,1)
solution=solve((eq1,eq2),(x,y))
print(solution)
```
